<b>Metrics</b>

The primary metric used will be accuracy, because that is what the Kaggle competition is benchmarked on.<br>

$$
\text{Accuracy} = \frac{\text{True Positives + True Negatives}}{\text{Total Instances}}
$$<br>

Precision will also be used, in case the business wants to limit false possitives.<br>

$$
\text{Precision} = \frac{\text{TP}}{\text{TP} + \text{FP}}
$$<br>
And recall will be used, in case the buesiness wants to limit false negatives.<br>

$$
\text{Recall} = \frac{\text{True Positives}}{\text{True Positives + False Negatives}}
$$<br>

We will also consider F1 score, which is the harmonic mean of the precision and recall that ranges between 0 and 1.


$$
\text{F1} = \frac{\text{2}}{\text{Recall}^{-1}+\text{precision}^{-1}}=\frac{\text{2TP}}{\text{2TP+FP+FN}}
$$<br>


These metrics will be used to measure the performance of models on predicting whether a client will pay back loan on time. Which metric is ultimatley used depends on the tradeoff between the risks of giving a loan to someone who won't repay it, or not givng a loan to someone who would repay it. We will also use a confusion matrix to get an overview these different types of errors. F1 is also included to balance the precision and recall scores due to the imblanced nature of the target variable in this dataset. 


## Machine Learning Pipelines (Trevor and Swapnali)

### <b>Baseline Pipeline</b>

A baseline pipeline is a pipeline that does not aim to produce the greatest accuracy or precision but is made for base level of performance. This baseline pipeline follows many of the same steps as any other pipeline but usally forgoes high levels of feature engineering or hyperparameter tuning. The reason for this is it will act as our goal post when creating feature pipelines, it is purely a comparison tool. That being said, it still must be competent in its construction. The steps in its creation are as indicated below. 
1.	Data Cleaning: The data will be extracted, loaded, and prepared. This includes dealing with missing values, performing any needed data transformations or data normalizations. In our case this will primarily consist of dealing with missing variables.  
2.	Feature Selection: Though a baseline it is still important to identify any important information based on the EDA when it comes to impactful features. 
3.	Model Selection: a baseline model that we believe would be appropriate such as in our case logistic regression or KNN and use this model as a starting point to train our data against. The model for the baseline should be simplistic. 
4.	Metrics Evaluation: Though it is a baseline model it must, as indicated before, be competent. We will verify through precision and accuracy of training, validation, and testing that there is nothing going awry within the pipeline (e.g. overfitting or underfitting).

### <b>Model Pipeline</b>

With a baseline pipeline established and an anchoring point made, we can focus on the construction of a better model pipeline. Since we will be editing the pipeline constructed above, we do not need to repeat data cleaning or any aspects of EDA. We will just follow the steps below to improve upon our baseline pipeline.

This could mean changing the scales of our features or normalizing. We may also choose to combine our features to create a new feature. 
2.	Model Selection and Optimization: It is at this point we will try different models, or even combining models. With this data we would see where KNN in combination with Nave Bayes could lead to an improved pipeline. 
3.	Hyperparameter Tuning: With a baseline established and a more complex model selected we can now tune hyperparameter in an effort to improve performance. 
4.	Metrics Evaluation: Not only do we still have our traditional metrics, but we can now evaluate metrics across different pipelines, including the baseline pipeline. The goal is to ensure we are not overfitting while still making improvements to our pipelines.
